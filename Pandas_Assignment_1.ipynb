**5.Which car manufacturers are having average co2 emissions above the average of all manufacturers?**

```python
# write your code here
#calculate the average of all manufacturers
#get the manufacturers having average co2 emissions above the average of all manufacturers
avg_Co2=data['CO2 Emissions(g/km)'].mean()
avg_Co2
avg_co2_manfactures=data.groupby('Make')['CO2 Emissions(g/km)'].mean()
avg_set=avg_co2_manfactures[avg_co2_manfactures>avg_Co2].sort_values(ascending=True)
avg_set
```

**9.Calculate the interquartile range for co2 emissions by small suvs.**
1. Step 1 : Filter the data for vehicle class small suv
2. step 2 : Calculate quantile 1 for co2 emission
3. step 3 : Calculare quantile 3 for co2 emmision
4. step 4 : calculate the IQR using step 2 and step 4.


```python
# write your code here
suv_data=data[data['Vehicle Class']=="SUV - SMALL"]
q1=suv_data['CO2 Emissions(g/km)'].quantile(0.25)
q3=suv_data['CO2 Emissions(g/km)'].quantile(0.75)
q5=q3-q1
q5
```
