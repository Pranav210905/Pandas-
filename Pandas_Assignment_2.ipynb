**5.Identify which car manufacturers have above-average fuel consumption compared to all manufacturers.**

```python
# write your code here
#calculate the average of all manufacturers
avg_fuel=data['Fuel Consumption Comb (L/100 km)'].mean()
avg_man_fuel=data.groupby('Make')['Fuel Consumption Comb (L/100 km)'].mean()
avg_set=avg_man_fuel[avg_man_fuel>avg_fuel]
avg_set
```

**7.Determine which combination of engine size and fuel type results in the highest average CO2 emissions.**

```python
# Group the data by 'Engine size' and 'Fuel Type', then calculate the mean of 'CO2 Emissions(g/km)' for each group.
# Sort the resulting mean values in descending order to show which combinations of engine size and fuel types have the highest average CO2 emissions.
dataVal=data.groupby(['Engine Size(L)','Fuel Type'])['CO2 Emissions(g/km)'].mean().sort_values(ascending=False)
dataVal.head(1)
```
